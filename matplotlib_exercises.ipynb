```python
#Combine the figures you created in the last step into one large figure with 4 subplots.
n_rows = 2
n_cols = 2

# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
# y = âˆšx
x = np.arange(0,50,1)
y = np.sqrt(x)

plt.plot(x,y)

# plot the second subplot
plt.subplot(n_rows, n_cols, 2)
# y = x^3
x = np.arange(0,10,1)
y = np.power(x,3)

plt.plot(x,y)

# plot the third subplot
plt.subplot(n_rows, n_cols, 3)
# y = 2^x
x = np.arange(0,10,1)
y = 2 **x

plt.plot(x,y)

# plot the fourth subplot
plt.subplot(n_rows, n_cols, 4)
# y = 1/(x+1)
x = np.arange(0,10,1)
y = 1/(x+1)

plt.plot(x,y)
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
plt.figure(figsize=(14, 6)) # (width, height)

# y = âˆšx
x = np.arange(0,5,1)
y = np.sqrt(x)

plt.plot(x,y, c='red', label='$y=âˆšx$')

# y = x^3
x = np.arange(0,5,1)
y = np.power(x,3)

plt.plot(x,y, c='orange', label='$y=x^3$')

# y = 2^x
x = np.arange(0,5,1)
y = 2 **x

plt.plot(x,y, c='black', label='$y = 2^x$')

# y = 1/(x+1)
x = np.arange(0,5,1)
y = 1/(x+1)

plt.plot(x,y, c='purple',label='$y = 1/(x+1)$')

plt.legend()
plt.title('Variables of x')


#plt.show()
```
